# Layerwise evaluation on MNIST

## MNIST Dataset Overview

The dataset contains 60,000 examples for training and 10,000 examples for testing. The digits have been size-normalized and centered in a fixed-size image (28x28 pixels) with values from 0 to 1. For simplicity, each image has been flattened and converted to a 1-D numpy array of 784 features (28*28).

![MNIST Dataset](http://neuralnetworksanddeeplearning.com/images/mnist_100_digits.png)

More info: http://yann.lecun.com/exdb/mnist/

```python
import tensorflow as tf
from tensorflow.python.tools import inspect_checkpoint
from tensorflow.python import pywrap_tensorflow

import numpy as np
import matplotlib.pyplot as plt

from matplotlib.backends.backend_pdf import PdfPages
```

```python
def load_variables(file_name):
    reader = pywrap_tensorflow.NewCheckpointReader(file_name)
    var_to_shape_map = reader.get_variable_to_shape_map()
    var_values = dict()
    for key in sorted(var_to_shape_map):
        print("tensor_name: ", key)
#         print(reader.get_tensor(key))
        var_values.update({key: reader.get_tensor(key)})
    return var_values
```

```python
# set checkpoint path

ckpt_path = '/home/popescu/repos/awp_network_quantization/trained_models/vgg_experiments/pruning/vgg_mnist_260919/model.ckpt-15407'
```

```python
variables = load_variables(ckpt_path)
```

```python
tensors_to_inspect = ['vgg_16/conv1/conv1_1/bias', 'vgg_16/conv1/conv1_1/weights', 
                      'vgg_16/conv1/conv1_2/bias', 'vgg_16/conv1/conv1_2/weights', 
                      'vgg_16/conv2/conv2_1/bias', 'vgg_16/conv2/conv2_1/weights', 
                      'vgg_16/conv2/conv2_2/bias', 'vgg_16/conv2/conv2_2/weights', 
                      'vgg_16/conv3/conv3_1/bias', 'vgg_16/conv3/conv3_1/weights', 
                      'vgg_16/conv3/conv3_2/bias', 'vgg_16/conv3/conv3_2/weights', 
                      'vgg_16/conv3/conv3_3/bias', 'vgg_16/conv3/conv3_3/weights', 
                      'vgg_16/conv4/conv4_1/bias', 'vgg_16/conv4/conv4_1/weights', 
                      'vgg_16/conv4/conv4_2/bias', 'vgg_16/conv4/conv4_2/weights', 
                      'vgg16/conv4/conv4_3/bias', 'vgg16/conv4/conv4_3/weights', 
                      'vgg_16/conv5/conv5_1/bias', 'vgg_16/conv5/conv5_1/weights', 
                      'vgg_16/conv5/conv5_2/bias', 'vgg_16/conv5/conv5_2/weights', 
                      'vgg_16/conv5/conv5_3/bias', 'vgg_16/conv5/conv5_3/weights', 
                      'vgg_16/fc6/bias', 'vgg_16/fc6/weights',
                      'vgg_16/fc7/bias', 'vgg_16/fc7/weights',
                     ]
```

```python
subplots_filename = './eval_results/vgg_subplots.pdf'
```

```python
with PdfPages(subplots_filename) as pdf:
    f,a = plt.subplots(15,2, figsize=(50,50))
    a = a.ravel()

    for idx,ax in enumerate(a):
        ax.hist(np.asarray(variables[tensors_to_inspect[idx]]).flatten())
        ax.set_title(tensors_to_inspect[idx])
    plt.tight_layout()
    
    pdf.savefig()
```

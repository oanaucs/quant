```python
# plot points
fig, ax = plt.subplots()

ax.grid()
poly = np.polyfit(pruning_data['compression_ratio'], pruning_data['accuracy_loss'],10)
x = np.linspace(pruning_data['compression_ratio'].min(), pruning_data['compression_ratio'].max(), 100)
p = np.poly1d(poly)
plt.plot(x, p(x), color='green')
plt.plot(pruning_data['compression_ratio'], pruning_data['accuracy_loss'], 'bo', color='green', label='pruning')

poly = np.polyfit(pruning_quant_data['compression_ratio'], pruning_quant_data['accuracy_loss'],10)
x = np.linspace(pruning_quant_data['compression_ratio'].min(), pruning_quant_data['compression_ratio'].max(), 100)
p = np.poly1d(poly)
plt.plot(x, p(x), color='red')
plt.plot(pruning_quant_data['compression_ratio'], pruning_quant_data['accuracy_loss'], 'bo', color='red', label='pruning+quantization')
plt.ylim((-0.1, 0.05))
plt.xlabel('model ratio')
plt.ylabel('accuracy loss')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

```python
# plot time step points
fig, ax = plt.subplots()

ax.grid()
poly = np.polyfit(pruning_quant_data['avg_quant_time'], pruning_quant_data['bits'],2)
x = np.linspace(pruning_quant_data['avg_quant_time'].min(), pruning_quant_data['avg_quant_time'].max(), 100)
p = np.poly1d(poly)
plt.plot(x, p(x))
plt.plot(pruning_quant_data['avg_quant_time'], pruning_quant_data['bits'], 'bo', color='darkblue', label='average step time')

poly2 = np.polyfit(pruning_quant_data['layer_quant_time'], pruning_quant_data['bits'], 2)
x2 = np.linspace(pruning_quant_data['layer_quant_time'].min(), pruning_quant_data['layer_quant_time'].max(), 100)
p2 = np.poly1d(poly2)
plt.plot(x2, p2(x2))
plt.plot(pruning_quant_data['layer_quant_time'], pruning_quant_data['bits'], 'bo', color='orange', label='layer quantization time')
plt.xlabel('time[s]')
plt.ylabel('bits')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```
